```python
import pandas as pd
df = pd.read_csv('portfolios.csv')
```

```python
column_means = df.iloc[:, 1:].mean(axis=0)
column_means_df = column_means.to_frame().transpose()
all_list = column_means_df.values.tolist()
import numpy as np
std_port = np.std(all_list)
all_data = np.array(all_list)[0]
print(f'The standard Deviation of returns of All portfolio is: {std_port}')
```

```python
print(all_data)
```

```python
GSPC_List = pd.read_csv('GSPC.csv')
date_list = GSPC_List['Date']
GSPC_returns = GSPC_List['Returns']
```

```python
std_GSPC = np.std(GSPC_returns)
print(std_GSPC)
```

```python
import matplotlib.pyplot as plt
weeks = range(len(date_list))
sector_title = 'All Sectors'
plt.figure(figsize=(12, 8))
plt.xlabel('Date',fontsize=16)
plt.ylabel('Returns, (%)',fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=18)
plt.grid(which='both', linestyle='--', linewidth=0.5)


plt.plot(date_list, all_data* 100, label = f'Returns of Portfolio in {sector_title}')
plt.plot(date_list, GSPC_returns*100, label = f'Returns of S&P 500 Factors')
plt.legend(fontsize=14)
plt.title('Returns of Portfolio in ' + sector_title , fontsize=18)

interval = 6 
plt.xticks(ticks=date_list[::interval], labels=date_list[::interval], rotation=45)
plt.savefig('Portfolio_Returns', bbox_inches='tight')
plt.show()
```
